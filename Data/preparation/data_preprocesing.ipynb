```python
# Hàm thống kê type
def statis(df: pd.DataFrame):
    df["type"] = df["question_idx"].apply(lambda x: x.split("_")[0])
    return df.groupby("type").count()["question_idx"]

# Hàm vẽ có thêm số
def draw_all_plots(datasets, titles = ['Train Set', 'Validation Set', 'Test Set']):
    
    fig, axs = plt.subplots(1,len(titles), figsize=(18, 5))  # 1 dòng, 3 biểu đồ
   
    for i, (series, title) in enumerate(zip(datasets, titles)):
        ax = axs[i]
        series.plot(kind='bar', color='skyblue', edgecolor='black', ax=ax)

        # Thêm giá trị trên đầu cột
        for j, v in enumerate(series.values):
            ax.text(j, v + max(series.values)*0.01, str(v), ha='center', va='bottom', fontweight='bold')

        ax.set_title(f"Số mẫu theo type - {title}")
        ax.set_ylabel("Số lượng")
        ax.set_xlabel("Type")
        ax.tick_params(axis='x', rotation=15)

    plt.tight_layout()
    plt.show()



# Gọi hàm vẽ toàn bộ
draw_all_plots([statis(train_ds), statis(valid_ds), statis(test_ds)])
```

```python
merged_train = pd.concat([train_ds, valid_ds])
merged_train.shape
```

```python
draw_all_plots([statis(merged_train), statis(test_ds)], ['Train Set', 'Test Set'])
```
